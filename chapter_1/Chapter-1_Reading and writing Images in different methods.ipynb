Method-4
- open using matplotlib

```python
# read the image using pillow and open the image using matplotlib
import matplotlib.pyplot as plt
from PIL import Image
image=Image.open('Urus-Greymatt.jpg')
plt.imshow(image)
```

```python
# read the image using cv2 and open by using matplotlib
import cv2
import matplotlib.pyplot as plt
image=cv2.imread('Urus-Greymatt.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```
