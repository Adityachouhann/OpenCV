```python
import matplotlib.pyplot as plt
import cv2
# read the image using open cv or using pillow 
image_bgr=cv2.imread(r"C:\Users\ajcho\OneDrive\Desktop\full stack data science and AI\BY SELF\OPEN-CV\chapter_2\Benjamin-Franklin.jpg")
#open the image using matplotlib
image_rgb= cv2.cvtColor(image_bgr,cv2.COLOR_BGR2RGB)
plt.subplot(1,2,1).imshow(image_bgr)
plt.subplot(1,2,2).imshow(image_rgb)
plt.show()
```

```python
#Translation matrix the first row moves pixels horizontally and the second row moves in vertically.
import numpy as np
M = np.float32([[1, 0, 250], [0, 1, 500]])
M

# 1 0  is horizontal axis
# 0 1  is vertical axis
```

```python
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape
[0]))
plt.imshow(shifted)
plt.show()
```

```python
#Changed the pixel directions to move image left and up
#[1, 0, -50] [row,column,number of pixels]
M = np.float32([[1, 0, -50],   # Horizontal move 
                [0, 1, -100]]) # Vertical move
shifted = cv2.warpAffine(image, 
                         M, 
                         (image.shape[1], 
                          image.shape[0]))
plt.imshow(shifted)
plt.show()
```

rotation 

```python
(height, width, channels) = image.shape
centre=(width//2, height//2)
centre
```

```python
M = cv2.getRotationMatrix2D(centre, 45, 1.0)
M
```

```python
M = cv2.getRotationMatrix2D(centre, 45, 1.0)
rotated = cv2.warpAffine(image, M, (width, height))
plt.imshow(rotated)
plt.show()
```

```python
M = cv2.getRotationMatrix2D(centre, 90, 1.0)
rotated = cv2.warpAffine(image, M, (width, height))
plt.imshow(rotated)
plt.show()
```

```python
M = cv2.getRotationMatrix2D(centre, 180, 1.0)
rotated = cv2.warpAffine(image, M, (width, height))
plt.imshow(rotated)
plt.show()
```

```python
M = cv2.getRotationMatrix2D(centre, -90, 1.0)
rotated = cv2.warpAffine(image, M, (width, height))
plt.imshow(rotated)
plt.show()
```
